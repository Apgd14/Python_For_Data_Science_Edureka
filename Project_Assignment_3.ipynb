```python
# Importing the necessary libraries 
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

print("Import Successful")
```

```python
# Set the working directory to where the dataset resides
os.chdir("F:\\Python_Edureka\\projects\\Project_3")

# Verify the set working directory
os.getcwd()
```

```python
# Importing train_data file into pandas Dataframe
train_data = pd.read_csv("train.csv")

# Display the first five rows of the train_data set
train_data.head()
```

```python
# Importing test_data file into pandas Dataframe
test_data = pd.read_csv("test.csv")

# Display the first five rows of the test_data set
test_data.head()
```

```python
# Label Encoding the categorical variables of the train_data dataset
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_data["species"] = le.fit_transform(train_data["species"])
train_data.head()
```

```python
# Creating X and y out of the train_data
X = train_data.iloc[:,2:].values
y = train_data.iloc[:,1].values
```

```python
# Splitting the data using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 5)
```

```python
# Applyong Random Forest Classifier for the train_data

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators = 100)
rf_model = rf_clf.fit(X_train, y_train)
```

```python
# Predicting the random forest model on the X_test

y_rf_pred = rf_model.predict(X_test)

# Checking the accuracy of the random forest model

from sklearn.metrics import accuracy_score
print("Random Forest Model Accuracy: ", accuracy_score(y_rf_pred, y_test))
```

### Fitting a Support Vector Machine Classifier Model

```python
# Fitting a Support Vector Machine for data

from sklearn import svm
svm_clf = svm.SVC()
model_svm = svm_clf.fit(X_train, y_train)
```

```python
# Predicting the svm model on the  X_test data

y_svm_pred = model_svm.predict(X_test)

# Checking the accuracy of the svm model

print("SVM accuracy:", accuracy_score(y_svm_pred, y_test))
```

### Fitting a Naive Bayes Model 

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb_model = gnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb_model.predict(X_test)
print("Naive_Bayes accuracy:", accuracy_score(y_pred_gnb, y_test))
```

### Fitting a Decision Tree Model

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
model_dt = dt.fit(X_train, y_train)
```

```python
y_pred_dt = model_dt.predict(X_test)
print("Decision_tree accuracy:", accuracy_score(y_pred_dt, y_test))
```
